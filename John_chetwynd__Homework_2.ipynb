```
# Write code below
# 1) What % of Netflix content is produced in the USA? You can calculate the numerator and denominator separately. (Hint: Use the new column you created in Section 1)
percentMadeInUsa = (producedInUsa / lengthOfDataSet) * 100 
print(percentMadeInUsa, "percent of the data set is produced in the USA")

# 2) List all the movies directed by M. Night Shyamalan that are available on Netflix.
NightShyamalan = netflixData.director.str.contains("Shyamalan",regex=False)
netflixData[NightShyamalan]

# 3)Which TV Shows produced in Japan that were released in 2006 are available on Netflix? 
#  FILTER DATA SET FOR TWO OR MORE CONDITIONS USING THIS SYNTAX
booleanIndex = (netflixData['country'] == "Japan") & (netflixData['release_year'] == 2006)
netflixData[booleanIndex]

# 4)How many movies and how many TV shows (reported separately) did Netflix add in the month of July 2019
netflixData.date_added.fillna("Unkown", inplace=True)
startsWithJuly = netflixData.date_added.str.startswith("July")
contains2019   = netflixData.date_added.str.contains("2019")
containsMovies = netflixData.type.str.contains("Movie")
containsTVShows= netflixData.type.str.contains("TV Show")
moviesInJuly2019  = netflixData[startsWithJuly & contains2019 & containsMovies]
TVShowsInJuly2019 = netflixData[startsWithJuly & contains2019 & containsTVShows]
print(len(moviesInJuly2019))
print(len(TVShowsInJuly2019))

# 5)List the Netflix content (movies/TV shows) about the FBI (description contains "FBI") chronologically by release year (most recent first)
netflixData.sort_values(by=['release_year'], inplace=True , ascending= False ) 
FBI = netflixData.description.str.contains("FBI")
netflixData[FBI]
```
